Zamiana wartoÅ›ci sÅ‚ownych na wartoÅ›ci binarne

```python
df.replace({"Gender": {"Male": 0, "Female": 1}}, inplace=True)
df.replace({"Customer Type": {"Loyal Customer": 1, "disloyal Customer": 0}}, inplace=True)
df.replace({"Type of Travel": {"Personal Travel": 0, "Business travel": 1}}, inplace=True)
df.replace({"satisfaction": {"satisfied": 1, "neutral or dissatisfied": 0}}, inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse_output=False)
feat = enc.fit_transform(df[["Class"]])
print(feat)
df["Business"] = feat[:,0]
df["Eco"] = feat[:,1]
df["Eco Plus"] = feat[:,2]
df = df.drop(columns=['Class'])
# df.head(10)
```

Przeskalowanie do wartoÅ›ci z zakresu 0-1

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
feat_to_scale = ["Age", "Flight Distance", "Inflight wifi service", "Departure/Arrival time convenient", "Ease of Online booking", "Gate location",'Food and drink','Online boarding','Seat comfort','Inflight entertainment','On-board service','Leg room service','Baggage handling','Checkin service','Inflight service','Cleanliness','Departure Delay in Minutes','Arrival Delay in Minutes', 'satisfaction']
feat = scaler.fit_transform(df[feat_to_scale])
for i,col in enumerate(feat_to_scale):
    df[col] = feat[:,i]
col_y = df.pop('satisfaction')
df['Satisfaction'] = col_y
df.head(n=10)
```

Trenowanie

```python
from sklearn.model_selection import train_test_split
import time
from sklearn.metrics import classification_report, confusion_matrix

df.shape
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test, = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def clf_score(nazwa_modelu, y_pred, czas_treningu, macierz_pomylek, y_test=y_test):
    y_t = pd.Series(y_test).replace({0: 'neutral or dissatisfied', 1: 'satisfied'})
    y_p = pd.Series(y_pred).replace({0: 'neutral or dissatisfied', 1: 'satisfied'})

    return {
        'Model': nazwa_modelu,
        'DokÅ‚adnoÅ›Ä‡': accuracy_score(y_t, y_p),
        'Precyzja (neutral or dissatisfied)': precision_score(y_t, y_p, pos_label='neutral or dissatisfied'),
        'Precyzja (satisfied)': precision_score(y_t, y_p, pos_label='satisfied'),
        'CzuÅ‚oÅ›Ä‡ (neutral or dissatisfied)': recall_score(y_t, y_p, pos_label='neutral or dissatisfied'),
        'CzuÅ‚oÅ›Ä‡ (satisfied)': recall_score(y_t, y_p, pos_label='satisfied'),
        'F1-score (neutral or dissatisfied)': f1_score(y_t, y_p, pos_label='satisfied'),
        'F1-score (satisfied)': f1_score(y_t, y_p, pos_label='satisfied'),
        'Czas treningu [s]': czas_treningu,
        'Macierz pomylek': macierz_pomylek
    }
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
# import matplotlib.pyplot as plt

n_modeli = len(wyniki)
n_cols = 2
n_rows = (n_modeli + 1) // n_cols

# siatka wykresÃ³w
fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, n_rows * 4))
axes = axes.flatten()

for i, wynik in enumerate(wyniki):
    ax = axes[i]
    cm = wynik['Macierz pomylek']
    model_name = wynik['Model']

    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['neutral or dissatisfied', 'satisfied'])
    disp.plot(ax=ax, cmap='Blues', colorbar=True)
    ax.set_title(f"Macierz pomyÅ‚ek â€“ {model_name}")
    ax.set_xlabel("Przewidywana klasa")
    ax.set_ylabel("Rzeczywista klasa")

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
